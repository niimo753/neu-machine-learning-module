```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
```

# Problem 4

```python
data = pd.read_csv("../data/diabetes.csv")
data
```

```python
# split the dataset
X_train, X_test, y_train, y_test = train_test_split(data.iloc[:, :-1], data.iloc[:, -1], test_size=0.1, random_state=42)

# define the classifer
dtree = DecisionTreeClassifier(criterion="entropy")

# fit the data
dtree.fit(X_train, y_train)

# prediction
y_pred = dtree.predict(X_test)

# accuracy score
print("Accuracy for test set:", accuracy_score(y_test, y_pred))
print("Accuracy for train set:", accuracy_score(y_train, dtree.predict(X_train)))
```

```python
fig, ax = plt.subplots(figsize=(20, 20))
plot_tree(dtree, fontsize=5, feature_names=list(X_train.columns), ax=ax, filled=True)
```
